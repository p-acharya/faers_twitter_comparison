```python
drug_query = """SELECT OUTCOME.PRIMARYID, OUTCOME.CASEID, OUTCOME.OUTC_COD, 
DRUG.ROLE_COD, DRUG.DRUGNAME, DRUG.VAL_VBM, DRUG.DECHAL, DRUG.RECHAL, DEMOGRAPHIC.CASEVERSION
FROM ((OUTCOME
INNER JOIN DRUG ON OUTCOME.PRIMARYID = DRUG.PRIMARYID AND OUTCOME.CASEID = DRUG.CASEID) 
INNER JOIN DEMOGRAPHIC OUTCOME.PRIMARYID = DEMOGRAPHIC.PRIMARYID AND OUTCOME.CASEID = DEMOGRAPHIC.CASEID)"""
```

```python
outcome_drug = pd.read_sql_query(drug_query, conn)
```
