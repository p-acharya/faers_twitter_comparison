```python
drug_query = """
SELECT
  OUTCOME.PRIMARYID,
  OUTCOME.CASEID,
  OUTCOME.OUTC_COD,
  DRUG.ROLE_COD,
  DRUG.DRUGNAME,
  DRUG.VAL_VBM,
  DRUG.DECHAL,
  DRUG.RECHAL,
  DEMOGRAPHIC.CASEVERSION,
  REACTION.PT,
  DEMOGRAPHIC.FDA_DT,
  DEMOGRAPHIC.AGE,
  DEMOGRAPHIC.AGE_COD,
  DEMOGRAPHIC.GNDR_COD,
  SOURCE.RPSR_COD
FROM
  (
    (
      (
        (
          (
            OUTCOME
            LEFT OUTER JOIN DRUG ON OUTCOME.PRIMARYID = DRUG.PRIMARYID
            AND OUTCOME.CASEID = DRUG.CASEID
          )
          LEFT OUTER JOIN DEMOGRAPHIC ON OUTCOME.PRIMARYID = DEMOGRAPHIC.PRIMARYID
          AND OUTCOME.CASEID = DEMOGRAPHIC.CASEID
        )
        LEFT OUTER JOIN REACTION ON OUTCOME.PRIMARYID = REACTION.PRIMARYID
        AND OUTCOME.CASEID = REACTION.CASEID
      )
      LEFT OUTER JOIN SOURCE ON OUTCOME.PRIMARYID = SOURCE.PRIMARYID
      AND OUTCOME.CASEID = SOURCE.CASEID
    )
  )

"""
```

```python
od = pd.read_sql_query(drug_query, conn)
```
